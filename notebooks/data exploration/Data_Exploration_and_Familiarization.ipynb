# **Combines the data**

Relationship.JPG

```python
merge_data = pd.merge(bureau, customer, on = 'cust_id', how = 'inner')
final_data = pd.merge(merge_data, loan, on = 'cust_id', how = 'inner')

final_data.head()
```

# **Explore the Categorical Columns**

```python
cat_cols = ['gender', 'marital_status', 'employment_status', 'residence_type', 'loan_purpose', 'loan_type', 'default']

fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()

for i, column in enumerate(cat_cols):
    sns.countplot(data=df, x=column, ax=axes[i], palette='Set2')
    axes[i].set_title(f"Distribution of {column.capitalize()}")

# Remove empty subplots (if any)
for j in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Summary of Findings**

The charts show the distribution of various categorical variables in the dataset.

* **Gender:** The majority of the dataset consists of males.
* **Marital Status:** A higher proportion of individuals are married.
* **Employment Status:** Most people are self-employed.
* **Residence Type:** The majority of individuals own their residences.
* **Loan Purpose:** Home and personal loans are the most common purposes.
* **Loan Type:** Secured loans are more prevalent than unsecured loans.
* **Default:** A significant majority of loans are not in default.

# **Explore the Distribution of the Numeric Columns**

```python
# List of numeric columns
num_cols = df.select_dtypes(include=['number']).columns

# Create a 3x3 grid for subplots
fig, axes = plt.subplots(8, 3, figsize=(18, 36))  # Adjust figsize as needed
axes = axes.flatten()  # Flatten the 3x3 grid for easier indexing

# Plot each numeric column
for i, column in enumerate(num_cols):
    sns.boxplot(data=df, y=column, ax=axes[i], palette='Set2')
    axes[i].set_title(f"Boxplot of {column.capitalize()}")

# Remove empty subplots (if any)
for j in range(len(num_cols), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
